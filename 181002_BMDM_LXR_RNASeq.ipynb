Read in RPKM files for plotting

```python
spret_rpkm = pd.read_csv('./expression/rna_exp_spret_rpkm.tsv', sep='\t', index_col=0)
mm10_gene = spret_rpkm['Annotation/Divergence'].str.split('|').str[0]
spret_rpkm['gene'] = mm10_gene
spret_rpkm.index.rename('refseqID', inplace=True)
spret_rpkm = spret_rpkm.reset_index().set_index('gene').drop(labels='refseqID', axis=1)
spret_rpkm_mat = spret_rpkm.iloc[:, 7:]

c57_rpkm = pd.read_csv('./expression/rna_exp_c57_rpkm.tsv', sep='\t', index_col=0)
mm10_gene = c57_rpkm['Annotation/Divergence'].str.split('|').str[0]
c57_rpkm['gene'] = mm10_gene
c57_rpkm.index.rename('refseqID', inplace=True)
c57_rpkm = c57_rpkm.reset_index().set_index('gene').drop(labels='refseqID', axis=1)
c57_rpkm_mat = c57_rpkm.iloc[:, 7:]


# merge frames
rpkm = c57_rpkm.merge(spret_rpkm_mat, left_index=True, right_index=True)

# filter genes with length < 250 bp
rpkm = rpkm.loc[rpkm.Length >= 250, :]
```

extract sample data from IDs

```python
samples = pd.Series(rpkm.iloc[:, 7:].columns.values).str.split('/').str[-1]
strain = samples.str.split('_').str[0]
cage = samples.str.split('_').str[5]
group = ['_'.join(i) for i in samples.str.split('_').str[3:5]]
strain_group = strain+'_'+group

samplesDF = pd.DataFrame([pd.Series(samples),
                          pd.Series(cage),
                          pd.Series(group),
                          pd.Series(strain),
                          pd.Series(strain_group)],
                          index=['sample', 'cage', 'group', 'strain', 'strain_group']).T
```

run PCA

```python
# Import packages for PCA analysis
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA # uses SVD

def label_point(x, y, val, ax, fontsize=12):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        # print(i, point)
        ax.text(point['x']+.02,
                point['y'],
                str(point['val']),
                fontsize=fontsize)
```

```python
rpkm_filt = rpkm.loc[(rpkm.iloc[:, 7:] > 4).all(1), :]
# standardize matrix, mean=0, std=1
tpm_std = StandardScaler().fit_transform(rpkm_filt.iloc[:, 7:])
# run PCA
pca = PCA(n_components=20)
tpm_pca = pca.fit_transform(tpm_std.T)
exp_var = pca.explained_variance_ratio_
#create df for plotting
tpm_pca_df = pd.DataFrame(tpm_pca,
                          columns=['PC'+str(i) for i in range(1,21)],
        index=(samplesDF.loc[:, 'cage'])).reset_index()
tpm_pca_df['strain'] = samplesDF['strain']
tpm_pca_df['strain_group'] = samplesDF['strain_group']
```

### Pull comparisons from get diff files

```python
def pull_comparision(diff_gene, comp):
    return diff_gene.loc[:, [comp + ' Log2 Fold Change',
                            comp + ' p-value',
                            comp + ' adj. p-value']]

def count_p_fc(diff, pval=0.05, fc=2):
    
    tmp = [((diff.adj_pval <= pval) & (diff.log2fc >= np.log2(fc))).value_counts(),
            ((diff.adj_pval <= pval) & (diff.log2fc <= np.log2(fc))).value_counts()]
    print('up 1 vs. 2:')
    print(tmp[0])
    print('down 1 vs. 2:')
    print(tmp[1])
    return pd.DataFrame(tmp, index=['up', 'down']).fillna(0).T
```

### Compare results between strains

```python
def get_diff_hypergeom_test(df, df2, pval=0.05, fc=1):
    '''Takes two pandas dataframes in the get diff format
    last three cols fc, pval, adj_p and returns a gene overlap
    for them at a given p value and fc, need same index'''
    from scipy.stats import hypergeom
    import matplotlib_venn
    
    log2fc = np.log2(fc)
    # get gene lists
    df_sig = [df.loc[(df.iloc[:, -3] <= -log2fc) & (df.iloc[:, -1] <= pval), :].index.tolist(),
               df.loc[(df.iloc[:, -3] >= log2fc) & (df.iloc[:, -1] <= pval), :].index.tolist()]
    
    df2_sig = [df2.loc[(df2.iloc[:, -3] <= -log2fc) & (df2.iloc[:, -1] <= pval), :].index.tolist(),
               df2.loc[(df2.iloc[:, -3] >= log2fc) & (df2.iloc[:, -1] <= pval), :].index.tolist()]
    p = []
    fig, axs = plt.subplots(1, 2, figsize=(10,10))
    overlap={}
    differences={}
    for i, test in [(0, 'up'), (1, 'down')]:
        # run hypergeom test
        M = df.shape[0] # rows = total genes
        n = len(df_sig[i]) # number of sig genes group 1
        N = len(df2_sig[i]) # number of sig genes in group 2
        x = len(set(df_sig[i]).intersection(df2_sig[i]))
        hypergeo = hypergeom(M, n, N)
        p.append(hypergeo.pmf(x))
        overlap[test] = set(df_sig[i]).intersection(df2_sig[i])
        print('working on', test, 'genes')
        print('total genes:', M)
        print('df sig genes:', n)
        print('df2 sig genes:', N)
        print('overlap:', x)
        print('hypergeometric p value:', p[i])
        matplotlib_venn.venn2([set(df_sig[i]), set(df2_sig[i])],
                              set_labels=['df1_genes', 'df2_genes'],
                              ax=axs[i])
        axs[i].set_title('genes '+test+', fc > '+str(fc)+', p < '+str(pval))
    # find genes with different regulation ie up in 1 and down in 2
    overlap['up_down'] = set(df_sig[0]).intersection(df2_sig[1])
    overlap['down_up'] = set(df_sig[1]).intersection(df2_sig[0])
    # find 'private' up and down genes
    differences['up_df1'] = set(df_sig[0]) - set(df2_sig[0])
    differences['down_df1'] = set(df_sig[1]) - set(df2_sig[1])
    differences['up_df2'] = set(df2_sig[0]) - set(df_sig[0])
    differences['down_df2'] = set(df2_sig[1]) - set(df_sig[1])
    
    return (p, overlap, differences)
```
